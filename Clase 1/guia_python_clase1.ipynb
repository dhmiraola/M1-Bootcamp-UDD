### If

#### El if es una sentencia condicional, este permite condicionar el flujo de un programa en diferentes caminos. Es como darle al computador la capacidad de distinguir varias opciones para actuar de forma distinta

```python
if True: # Los 2 puntos significa que esta "indentado", es decir que est dentro de la condicin
    print("se cumple la condicin")
    print("Tambin se muestra este print")
```

```python
# Intenta hacer un if cambiando el True a False y comprueba
if False: 
    print("se cumple la condicin")
    print("Tambin se muestra este print")
```

```python
a = 10

if a == 2: # Se coloca con 2 signos igual
    print("a vale 2")
if a == 10:
    print("a vale 10") 

# El resultado solo reconocer el segundo ya que es el que cumple la condicin
```

```python
a = 5
b = 10

if a == 5:
    print("a vale", a)
if b == 10:
        print("y b vale",b)

```

```python
n = 0

if n == 0:
    print("n vale 0")
else:
    print("n vale", n)
    
# La sentencia else (sino) se encadena a un if para indicar el caso contrario cuando NO se cumple la condicin
```

```python
n = float(input("Introduce un nmero: "))

if n == 0:
    print("n vale 0")
elif n >= 5:
    print("n vale 5")
elif n >=  7:
    print("n vale 7")
else:
    print("n vale ",n)

# la sentencia elif muestra por pantalla la condicin que cumple n
```

### Bucle While

#### Bucle o Iterar: Es realizar una accin varias veces
#### Ejemplo prctico: imagine que tiene una base de datos con un montn de registros y necesita encontrar uno en especial para consultarlo, para encontrar el elemento el computador debe recorrer entre todos los registros de la base de datos y compararlos hasta encontrar el que se busca

#### La clave del while es que a partir de una expresin relacional o lgica, de la misma forma que trabaja la sentencia if es capax de repetir un bloque de instrucciones siempre que este devuelva True, nosotros tenemos que encargarnos de cambiar esta condicin a False para salir de la iteracin o del bucle, sino estaramos en un bucle infinito

```python
c = 0

while c <=5:
    c+=1
    print("c vale ",c)
    
# Le estamos pidiendo al bucle while que mientras la c sea mas pequea que 5, esta me incremente el valor de c en 1
```

```python
c = 0

while c <=5:
    c+=1
    print("c vale ",c)
else:
    print("Se ha completado toda la iteracin y c vale ",c)
    
# se puede utilizar un else que se cumplir una vez cuando la condicin no se cumpla
```

```python
# Sentencia break

c = 0
while c <= 5:
    c+=1
    if (c==4):
        print("Rompemos el bucle cuando c vale", c)
        break
    print("c vale",c)
else:
    print("Se ha completado toda la iteracin y c vale", c)
```

#### Ejercicio: Cree un men interactivo que tenga 3 instrucciones ENTRAR, SALIR y COMANDO DESCONOCIDO

```python
print("Bienvenido")
while(True):
    print("""Elige una opcin: 
    1) Entrar
    2) Salir""")
    opcion = input()
    if opcion == "1":
        print("Entrando al sistema")
    elif opcion == "2":
        print("Saliendo del sistema")
        break
    else:
        print("Comando desconocido")
```

#### Ejercicio: Cree un men interactivo que SALUDE, SUME 2 NMEROS y SALIR 

```python
print("Bienvenido al men interactivo")
while(True):
    print("""Qu quieres hacer? Escribe una opcin
    1) Saludar
    2) Sumar dos nmeros
    3) Salir""")
    opcion = input()
    if opcion == '1':
        print("Hola, espero que te lo ests pasando bien")
    elif opcion == '2':
        n1 = float(input("Introduce el primer nmero: "))
        n2 = float(input("Introduce el segundo nmero: "))
        print("El resultado de la suma es: ",n1+n2)
    elif opcion =='3':
        print("Hasta luego! Ha sido un placer ayudarte")
        break
    else:
        print("Comando desconocido, vuelve a intentarlo")
        
```
