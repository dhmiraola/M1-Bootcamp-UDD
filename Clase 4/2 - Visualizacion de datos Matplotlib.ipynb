```python
import matplotlib.pyplot as plt
```

# Histograma

```python
plt.figure(figsize=(8,9))
plt.hist(data['age'], bins=20, color='lightgreen', edgecolor='black')
plt.title("Distribuci√≥n de edad")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

# Torta

```python
plt.pie(data['sex'].value_counts(),
        labels=data['sex'].value_counts().index,
        autopct='%1.1f%%',
        colors=['lightblue', 'pink'])

plt.title("Proporci√≥n de hombres y mujeres")
plt.axis('equal')  # c√≠rculo perfecto
plt.show()
```

# Dispersion

```python
plt.figure(figsize=(8,9))
plt.scatter(data['age'], data['fare'], alpha=0.5, color='orange')
plt.title("Edad vs Tarifa pagada")
plt.xlabel("Edad")
plt.ylabel("Tarifa")
plt.grid(True)
plt.show()
```

# Boxplot

```python

plt.figure(figsize=(8,4))
plt.boxplot(data['age'].dropna()) # el dropna sirve para eliminar las columnas NaN
plt.title('Boxplot de Edad')
plt.show()
```
